```python
import numpy as np
import pandas as pd

path = r'C:\Users\91999\Desktop\Project trial\Dataset.json'
data = pd.read_json(path, lines= True,orient='columns')
data
```

```python
#dropping column
data.drop(['extras'],axis = 1,inplace = True)
data
```

```python
for i in range(0,len(data)):
    if data.annotation[i]['label'][0] == '1':
        data.annotation[i] = 1
    else:
        data.annotation[i] = 0
        
```

```python
data
```

```python
import string
string.punctuation
```

```python
#removing punctuations

def remove_punctuation(data):
    data_nopunct = "".join([c for c in data if c not in string.punctuation])
    return data_nopunct
```

```python
data['msg_clean'] = data['content'].apply(lambda x: remove_punctuation(x))
data
```

```python
#tokenization

import re
def tokenize(data):
    tokens = re.split('\W+',data)
    return tokens

data['tokenized'] = data['msg_clean'].apply(lambda x: tokenize(x.lower()))
data
```

```python
#removing stopwords

import nltk
stopwords = nltk.corpus.stopwords.words('english')
stopwords[0:10]
```

```python
def remove_stopwords(tokenized):
    token_clean = [word for word in tokenized if word not in stopwords]
    return token_clean

data['msg_no_sw'] = data['tokenized'].apply(lambda x: remove_stopwords(x))
data
```

```python
#stemming

from nltk.stem import PorterStemmer 
ps = PorterStemmer()
```

```python
def stemming(tokenized):
    text = [ps.stem(word) for word in tokenized]
    return text

data['msg_stemmed'] = data['msg_no_sw'].apply(lambda x: stemming(x))
data
```
